```python
col_list = ["Quantity", "UnitPrice"]
df = pd.read_csv("C:/Users/HP/Downloads/1985_3404_bundle_archive/data.csv", usecols=col_list)
df.describe()
```

```python
# Separating Data
x = df.values[:,0:7]
print(len(x))
```

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
x = min_max_scaler.fit_transform(x)
print(x)
```

```python
# Train Data
train_data = x[0:3000]
print(len(train_data))

# Test Data
test_data = x[3000:50000]
print(len(test_data))
```

```python
# Separating Labels
y = df.values[0:50000]
print(y.shape)


```

```python
from keras.utils import to_categorical

y = to_categorical(y)
print(y.shape)


```

```python

train_label = y[0:3000,]

test_label = y[3000:50000,]
print(train_label.shape)
print(test_label.shape)
```

```python
from keras import models
from keras import layers
```

```python
def baseline_model():
  network = models.Sequential()
  network.add(layers.Dense(11, activation='relu', input_dim=1))
  network.add(layers.Dense(15, activation='relu'))
  # network.add(layers.Dense(8, activation='relu'))
  # network.add(layers.Dense(4, activation='relu'))
  network.add(layers.Dense(10, activation='softmax'))
  network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
  return network
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
estimator = KerasClassifier(build_fn=baseline_model, epochs=100, batch_size=10, verbose=2)
kfold = KFold(n_splits=5, shuffle=True)
```

```python
results = cross_val_score(estimator, test_data, test_label, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
